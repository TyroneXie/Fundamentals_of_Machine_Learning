### 13、时间序列

```python
df_time = pd.read_csv(r'C:/Users/Administrator/Desktop/time_serial.csv',encoding='gbk')
df_time.head()
```

```python
# 处理时间字段：把datetime域切成 日期 和 时间 两部分
temp = pd.DatetimeIndex(df_time['datetime'])
df_time['date'] = temp.date
df_time['time'] = temp.time
df_time.head(4)
```

```python
# 设定hour这个小时字段
df_time['hour'] = pd.to_datetime(df_time.time, format="%H:%M:%S")
df_time['hour'] = pd.Index(df_time['hour']).hour
df_time.head(4)
```

```python
# 产出一个星期几的类别型变量
df_time['dayofweek'] = pd.DatetimeIndex(df_time.date).dayofweek
df_time.head(4)
```

```python
# 对时间类特征处理，产出一个时间长度变量
df_time['dateDays'] = (df_time.date - df_time.date[0]).astype('timedelta64[D]')
df_time.tail(4)
```

```python
# 统计一周各天注册人数
day_count = df_time.groupby('dayofweek')
day_count['registered'].sum().reset_index()
```

```python
# 统计是否为周末
df_time['Saturday']=0
df_time.Saturday[df_time.dayofweek==5]=1

df_time['Sunday']=0
df_time.Sunday[df_time.dayofweek==6]=1

df_time.head(6)
```
